### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data to plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

Note: Also add title, xlabel, and ylabel to the plot.

#### A scatter plot is a graph in which the values of two variables are plotted along two axes. The pattern of the resulting points reveals any correlation present between the variables.

````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0,2 ,50)
y = 3 + np.random.normal(0 ,2 ,len(x))

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
````

Use the following data:

#### The subplot() function in Matplotlib allows you to create multiple plots in one figure. It's useful when you want to compare different plots side by side or when you want to create subplots with different sizes or layouts.

````python
x1 = np.array([0,1,2,3,4,5])
y1 = np.array([0,100,200,300,400,500])

x2 = np.array([0,1,2,3,4,5])
y2 = np.array([50,20,40,20,60,70])

x3 = np.array([0,1,2,3,4,5])
y3 = np.array([10 ,20 ,30 ,40 ,50 ,60])

x4 = np.array([0 ,1 ,2 ,3 ,4 ,5])
y4 = np.array([200 ,350 ,250 ,550 ,450 ,150])

fig,a=plt.subplots(2 ,2)

a[0][0].plot(x1,y1)
a[0][0].set_title('Line Plot - 1')
a[0][1].plot(x2,y2)
a[0][1].set_title('Line Plot - 2')
a[1][0].plot(x3,y3)
a[1][0].set_title('Line Plot - 3')
a[1][1].plot(x4,y4)
a[1][1].set_title('Line Plot - 4')

plt.show()
````

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

A bar plot is a graph that represents the category of data with rectangular bars with lengths and heights that are proportional to the values they represent. It is used to compare different categories of data. 

In Matplotlib, plotting a bar plot is as easy as calling the bar() function on the PyPlot instance and passing in the categorical and numerical variables that we'd like to visualize. 

````python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.title("Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

plt.barh(company, profit)
plt.title("Horizontal Bar Plot")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
````

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot is a graphical representation of statistical data based on the minimum, first quartile, median, third quartile, and maximum. It is used to display a summary of a set of data. The box plot in Matplotlib library is created with the help of boxplot() function. 

````python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig = plt.figure(figsize =(10, 7))

# Creating axes instance
ax = fig.add_subplot(111)

# Creating plot
bp = ax.boxplot(data)

plt.title('Box plot')
plt.xlabel('Data')
plt.ylabel('Values')

# Removing top axes and right axes ticks
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()

plt.show()
````
